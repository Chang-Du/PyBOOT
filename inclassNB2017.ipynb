```python
def pr(x, y=2.0, z=3.0):
    '''
    multiplying the arguments
    Arguments:
        x: float
        y: float (optional, default 2.0)
    return:
        products of the arguments
    '''
    return x * y *z
```

```python
def extractEvenReverse(dataAr):
    '''Extracts even values from array and reverses it
    Arguments:
    dataAr: numpy array
    '''
    print (type(dataAr))
    if not isinstance(dataAr, np.ndarray):
        print ("you should have passed and array")
        return -1
    return dataAr[(dataAr % 2 == 0)][::-1]
    
```
